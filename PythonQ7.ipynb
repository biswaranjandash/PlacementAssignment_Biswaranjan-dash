```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
data = pd.read_csv('/content/output (3).csv')

# Extract the year from the 'year' column, handle out-of-bounds values
data['year'] = pd.to_datetime(data['year'], errors='coerce').dt.year

# Get all the Earth meteorites that fell before the year 2000
meteorites_before_2000 = data[data['year'] < 2000]

# Plot the count of meteorites by year before 2000
meteorites_before_2000['year'].value_counts().sort_index().plot(kind='bar')
plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Number of Earth Meteorites that Fell before 2000')
plt.show()

# Get all the earth meteorite coordinates that fell before the year 1970
coordinates_before_1970 = data[data['year'] < 1970]['geolocation']

# Extract latitude and longitude values
latitude = coordinates_before_1970.str.extract(r'\[(.*),')[0].astype(float)
longitude = coordinates_before_1970.str.extract(r',(.*)\]')[0].astype(float)

# Plot the earth meteorite coordinates on a scatter plot
plt.scatter(longitude, latitude)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Earth Meteorite Coordinates (Before 1970)')
plt.show()

# Get all the earth meteorites with a mass greater than 10000 kg
mass_greater_than_10000 = data[data['mass'] > 10000]

# Plot the histogram of earth meteorite masses
mass_greater_than_10000['mass'].plot(kind='hist', bins=10)
plt.xlabel('Mass (g)')
plt.ylabel('Count')
plt.title('Distribution of Earth Meteorite Masses (> 10000 kg)')
plt.show()
```
